```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

```python
URL = 'https://www.pdga.com/tour/event/55582'

response = requests.get(URL)
response.status_code
```

```python
response.text ##call URL html and show as text
soup = bs(response.text) ##soupify response
print(soup.prettify()) ##print in a more readable form
```

```python
##loop to iterate over event urls
event = 55582

while event != 55596 :
    url = f"https://www.pdga.com/tour/event/{event}"
    print(url)
    event = event+1
```

```python
event = 55581  ##first event id
results_big = []  ##empty list to be filled by loop

##loop to iterate over event urls
while event != 55596 :
    event = event+1
    url = f"https://www.pdga.com/tour/event/{event}"
    response = requests.get(url)
    soup = bs(response.text)

    
#for loop to pull specific tables from pdga pages    
    for y in soup:
        y = pd.read_html(str(soup))
        df = y[1]
        results_big.append(df)
        print(event)
        break
    
results_big
        
```
