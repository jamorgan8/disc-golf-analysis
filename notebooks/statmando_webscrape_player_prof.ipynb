## End of Prelim. Scrape

## Format list

```python
names = mpo['Name']
names = pd.DataFrame(names)
names = names.reset_index()
names = names.drop(labels='index', axis=1)
names
```

```python
##replace spaces with hyphens
names['Name'] = names['Name'].str.replace(' ', '-')
names['Name'] = names['Name'].str.lower()
#names = names['Name'].tolist()
#names
```

```python
##initialize empty list to populate with formatted names
pnames = []

##loop to drop diacritical (unicode) characters to match url for scrape
for x in names['Name']:
    string_encode = x.encode("ascii", "ignore")
    string_decode = string_encode.decode()
    pnames.append(string_decode)
pnames
```
