```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

```python
URL = 'https://statmando.com/stats/career-dgpt-stats-mpo'

response = requests.get(URL)
```

```python
response.status_code
```

```python
response.text ##call URL html and show as text
soup = bs(response.text) ##soupify response
print(soup.prettify()) ##print in a more readable form
```

```python
soup.find('table', attrs={'id':'stats'})
```

```python
table_html = str(soup.find('table', attrs={'id':'stats'}))

HTML(table_html)
```

```python
player_stats_career = pd.read_html(str(soup.find('table', attrs={'id':'stats'})))[0]
player_stats_career
```

```python
player_stats_career.to_csv('../data/dpgt-mpo-career-stats.csv')
```
