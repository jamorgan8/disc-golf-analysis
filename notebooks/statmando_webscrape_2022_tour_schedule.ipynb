```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

```python
URL = 'https://statmando.com/stats/tour-schedule-2022'
response = requests.get(URL)
response.status_code
```

```python
response.text ##call URL html and show as text
soup = bs(response.text) ##soupify response
print(soup.prettify()) ##print in a more readable form
```

```python
soup.findAll('table')
```

```python
soup.find('table', attrs={'id':'stats'})
```

```python
table_html = str(soup.find('table', attrs={'id':'stats'}))

HTML(table_html)
```

```python
tour_schedule_2022 = pd.read_html(str(soup.find('table', attrs={'id':'stats'})))[0]
tour_schedule_2022
```

```python
tour_schedule_2022.to_csv('../data/tour-schedule-2022.csv')
```
