```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

```python
URL = 'https://statmando.com/players/'

response = requests.get(URL)
```

```python
response.status_code
```

```python
response.text ##call URL html and show as text
soup = bs(response.text) ##soupify response
print(soup.prettify()) ##print in a more readable form
```

```python
soup.findAll('table')
```

```python
soup.find('table', attrs={'id':'players'})
```

```python
table_html = str(soup.find('table', attrs={'id':'players'}))

HTML(table_html)
```

```python
player_stats_career = pd.read_html(str(soup.find('table', attrs={'id':'players'})))[0]
player_stats_career
```

```python
player_stats_career = pd.DataFrame(player_stats_career)
player_stats_career = player_stats_career.loc[['Division'] == 'MPO']
player_stats_career
```

```python
player_stats_career.to_csv('../data/player-stats-career.csv')
```
