```python
import requests
from bs4 import BeautifulSoup as bs
from IPython.core.display import HTML
import pandas as pd
```

```python
URL = 'https://www.pdga.com/tour/event/55580'

response = requests.get(URL)
response.status_code
```

```python
response.text ##call URL html and show as text
soup = bs(response.text) ##soupify response
print(soup.prettify()) ##print in a more readable form
```

```python
df = pd.read_html(str(soup)) ##using this, will automatically pull constructs labelled as tables into pandas DFs
df
```

```python
las_vegas_open = df[1] ##this table contains MPO results for las vegas challenge--- 'Unamed' columns are the round rating
las_vegas_open
```

```python
las_vegas_open.to_csv('../data/las_vegas_challenge.csv')
```
